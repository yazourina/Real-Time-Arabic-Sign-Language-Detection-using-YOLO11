## 1. Install & Import Ultralytics

```python
#!pip install ultralytics

import ultralytics
from ultralytics import YOLO
print("Ultralytics version:", ultralytics.__version__)
ultralytics.checks()
!yolo checks
```

```python
#Displays information about your GPU hardware and usage.
!nvidia-smi
```

# 2. Train YOLO11n on your dataset
you can also try yolo11s.pt, yolo11m.pt, etc. later for higher accuracy

```python

model = YOLO("yolo11n.pt")

model.train(
    data="data\data.yaml",  
    epochs=30,
    imgsz=640,
    batch=32,
    name="arasl_yolo11n"
)
```

# 3. Validate model performance

```python
metrics = model.val(data="data\data.yaml")
print(metrics)
```

# 4. Run inference on validation images

```python
results = model.predict(
    source= r"data\valid\images",   
    save=True,
    conf=0.25
)
```

# 5. Export model for deployment
like ONNX, TorchScript, CoreML, etc.

you can also use: "torchscript", "tflite", "engine"

```python
model.export(format="torchscript") 
```
